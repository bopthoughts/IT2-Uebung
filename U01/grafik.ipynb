3-D-Grafik

```python
import numpy as np
import matplotlib.pyplot as plt
from Arrows3d import install_extensions, plotpoint, plotvec, autoscale
install_extensions()
%matplotlib widget
```

```python
# Nullvektor und Einheitsvektoren
nv = np.array([[0, 0, 0]]).T
evx = np.array([[1, 0, 0]]).T
evy = np.array([[0, 1, 0]]).T
evz = np.array([[0, 0, 1]]).T
```

```python
# Trivialfall: Die Fußpunkte der drei Stäbe bilden eine gleichseitiges Dreieck mit Kantenlänge a
a = 4
h = np.sqrt(3)/2 * a
display(a, h)
P = np.array([[0, h/3, 5]]).T
P1 = np.array([[-a/2, 0, 0]]).T
P2 = np.array([[a/2, 0, 0]]).T
P3 = np.array([[0, h, 0]]).T

# Punkt P und äußere Kraft
F = np.array([[0, 0, -5]]).T
```

```python
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, projection='3d')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

plotpoint(ax, P, marker="o" )
plotpoint(ax, P1, marker="o", color='red')
plotpoint(ax, P2, marker="o", color='blue' )
plotpoint(ax, P3, marker="o", color='green' )

# plotvec(ax, nv, evx*5, color='blue', linestyle='dashdot')
# plotvec(ax, nv, evy*5, color='red')
# plotvec(ax, nv, evz*5, color='green')

plotvec(ax, P1, P, color='red', lw=2)
plotvec(ax, P2, P, color='blue', lw=2)
plotvec(ax, P3, P, color='green', lw=2)
plotvec(ax, P-F, P, lw=5)
#ax.autoscale(True)
fig.tight_layout()
autoscale(ax, 1.2)
```
